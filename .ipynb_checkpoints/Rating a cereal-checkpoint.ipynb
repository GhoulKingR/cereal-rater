```python
cereals.corr()
```

```python
# iterate the correlation table to get high correlation in the independent variables
for column in cereals.corr():
    
    for row in cereals.corr()[column]:
        if (row != 1) and ((row >= 0.8) or (row <= -0.8)):
            print( column )
            print( row, '\n' )
```

```python
del cereals['potass']
```

```python
# collect all the keys that are not of type 'object'
keys = []
for key in cereals:
    if cereals[key].dtype != 'object':
        if key != 'rating':
            keys.append(key)
```

# Create a model to rate a cereal

```python
# split the model into training and testing
from sklearn.model_selection import train_test_split

cereals_train, cereals_test = train_test_split(cereals, test_size=0.2)
```

```python
# create a linear regression model to use get calories
from sklearn.linear_model import LinearRegression

model = LinearRegression()
x_train = cereals_train[ keys ]
y_train = cereals_train[ 'rating' ]
model.fit(x_train, y_train)
```

```python
# create a new linear regression model to rate
from sklearn.linear_model import LinearRegression

full_model = LinearRegression()
x_full = cereals[ keys ]
y_full = cereals[ 'rating' ]
full_model.fit(x_full, y_full)
```

```python
# assess the accuracy of the model
x_test = cereals_test[keys]
y_test = cereals_test['rating']
print( 'model accuracy:', str(int(model.score(x_test, y_test)*100)),'%' )
```

```python
# saving the model for later use
import joblib
```

```python
file = "cereal_rater.mdl"
joblib.dump(full_model, file)
```
